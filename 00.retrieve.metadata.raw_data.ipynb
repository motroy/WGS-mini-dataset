```python
import pysradb as sra
from pysradb.sraweb import SRAweb
import os
```

```python
db = SRAweb()
```

```python
df = db.sra_metadata("PRJNA429440", detailed=True)
df
```

```python
df_contig_subset = df.loc[df["instrument_model_desc"].str.contains("ILLUMINA")]
df_contig_subset
```

```python
for SRR in df_contig_subset["run_accession"]:
    cmd="fastq-dl --accession %s --provider ENA --outdir PRJNA429440.Illumina/%s" % (SRR,SRR)
    os.system(cmd)
```
